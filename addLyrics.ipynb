```python
albumURLs = ["https://genius.com/albums/Doja-cat/Planet-her",
            #  "https://genius.com/albums/Doja-cat/Planet-her-deluxe",
             "https://genius.com/albums/Doja-cat/Streets-remixes",
            #  "https://genius.com/albums/Doja-cat/Hot-pink-japan-version",
             "https://genius.com/albums/Doja-cat/Hot-pink",
            #  "https://genius.com/albums/Doja-cat/Amala-deluxe-version",
             "https://genius.com/albums/Doja-cat/Amala",
             "https://genius.com/albums/Doja-cat/Purrr-ep"]
```

```python
from bs4 import BeautifulSoup
import lxml
import requests
```

```python
lyrics_links_titles = []
for _ in range(len(albumURLs)):

    link = albumURLs[_]
    html = requests.get(link).text
    soup = BeautifulSoup(html, 'lxml')

    s = soup.find_all("div", class_ = "chart_row-content")

    lyrics_links = [x.find('a', href = True)['href'] for x in s]

    for x in range(len(lyrics_links)):
        l = lyrics_links[x]
        lyrics_links_titles.append(l)
    
```

```python
lyrics_links_titles
```

```python
file_names = []
for a in lyrics_links_titles: 
    file = r"" + a[28:] + ".txt"
    file_names.append(file)
file_names
```

```python
for i in range(0,len(file_names)):
    t = tracks[i]
    a = df[df["Track Title"] == t]["Album Name"].iloc[0]
    y = df[df["Track Title"] == t]["Year Released"].iloc[0]
    data = {"Track Number": [i+1]*len(file_names),
            "Track Title": [t]*len(file_names),
            "Album Name": [a]*len(file_names),
            "Year Released": [y]*len(file_names)}
    df2 = pd.DataFrame(data)
    print(df2)
    
```

```python
for f in range(len(file_names)):
    file_name = file_names[f]
    with open(file_name, "rb") as file:
        x = file.readlines()
        print(len(x))
        print(x)
        file.close()
```

```python
for i in range(0,len(file_names)):
    file_name = file_names[i]
    t = tracks[i]
    a = df[df["Track Title"] == t]["Album Name"].iloc[0]
    y = df[df["Track Title"] == t]["Year Released"].iloc[0]
    with open(file_name, 'rb') as file:
        x = file.readlines()
        length = len(x)
        data = {"Track Number": [i+1]*length,
                "Track Title": [t]*length,
                "Album Name": [a]*length,
                "Year Released": [y]*length,
                "Lyric": x}
        df2 = pd.DataFrame(data)
        file_name_2 = r"" + file_name[:-4] + ".csv"
        df2.to_csv(file_name_2)
```
